```python
import sys
import datetime

val = 5
print(str(type(val)))
print('Version of Python: ' + sys.version)

def createDatetimeObject(date_str, time_str):
  datetime_object = datetime.datetime(int(date_str[0:4]), int(date_str[5:7]), int(date_str[8:10]), int(time_str[0:2]), int(time_str[3:5]), int(time_str[6:8]))
  return datetime_object


def areOneHourApart(datetime_object1, datetime_object2):
  delta = abs(datetime_object2 - datetime_object1)

  if delta.days == 0 and delta.seconds == 3600:
    return 1
  else:
    return 0


def main():
  date_array = []
  time_array = []
  generation_array = []
  
  date_array.append('2015-11-25')  #0
  date_array.append('2015-11-25')  #1
  date_array.append('2015-11-25')  #2
  date_array.append('2015-11-25')  #3
  date_array.append('2015-11-25')  #4
  date_array.append('2015-11-25')  #5
  date_array.append('2015-11-25')  #6
  date_array.append('2015-11-26')  #7
  date_array.append('2015-11-26')  #8

  time_array.append('05:00:00')    #0
  time_array.append('07:00:00')    #1
  time_array.append('09:00:00')    #2
  time_array.append('12:00:00')    #3
  time_array.append('13:00:00')    #4
  time_array.append('14:00:00')    #5
  time_array.append('23:00:00')    #6
  time_array.append('00:00:00')    #7
  time_array.append('02:00:00')    #8
  
  generation_array.append(['2015-11-25', '05:00:00'])    #0
  generation_array.append(['2015-11-25', '07:00:00'])    #1
  generation_array.append(['2015-11-25', '09:00:00'])    #2
  generation_array.append(['2015-11-25', '12:00:00'])    #3
  generation_array.append(['2015-11-25', '13:00:00'])    #4
  generation_array.append(['2015-11-25', '14:00:00'])    #5
  generation_array.append(['2015-11-25', '23:00:00'])    #6
  generation_array.append(['2015-11-26', '00:00:00'])    #7
  generation_array.append(['2015-11-26', '02:00:00'])    #8
    
  print(generation_array)
  print(generation_array.pop())
  print(generation_array)
  noOfFeatures = 3
  noOfRecords = len(date_array)
  left = 0
  right = left + noOfFeatures
  previousOk = 0
  previousDatetime = createDatetimeObject(date_array[left], time_array[left])
  while right <= noOfRecords:
    ok = 1
    print(str(left) + " " + str(right - 1))
    for index in range(left + 1, right):
      currentDatetime = createDatetimeObject(date_array[index], time_array[index])
    
      if not areOneHourApart(previousDatetime, currentDatetime):
        ok = 0
        break

      previousDatetime = currentDatetime

    if ok:
      for index in range(left, right):
        sys.stdout.write(date_array[index] + ", " + time_array[index])
        if index < right - 1:
          sys.stdout.write(", ")
        else:
          nextDatetime = currentDatetime + datetime.timedelta(hours=1)
          nextDate = nextDatetime.date()
          nextTime = nextDatetime.time()
          sys.stdout.write(" are features to predict " + str(nextDate) + " " + str(nextTime))

      sys.stdout.write("\n")

      left += 1
      right += 1
      previousOk = 1
    else:
      if previousOk:
        left = right - 1
        right = left + noOfFeatures
      else:
        left += 1
        right += 1
      previousOk = 0
        
    previousDatetime = createDatetimeObject(date_array[left], time_array[left])
        

main()
```
